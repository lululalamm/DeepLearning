```python
import re
```

```python
base = "celebA/"
```

```python
anno_path = os.path.join(base,"Anno/list_attr_celeba.txt")
img_base = os.path.join(base,"Img/img_celeba/")
```

```python
check_class={
    'beard':['No_Beard','Goatee','Mustache'],
    'smile':['Smiling'],
    'eyeglasses':['Eyeglasses'],
    'lipstick':['Wearing_Lipstick'],
    'hat':['Wearing_Hat'],
    'earrings':['Wearing_Earrings'],
    'necklace':['Wearing_Necklace'],
    'necktie':['Wearing_Necktie']
}
```

```python
check_class_index={}
annotations = {}
check_dict={}

with open(anno_path,'r') as f:
    # Get the total number of annotation
    num_annotation = int(f.readline()[:-1])
    
    # Get classes names
    classes = re.findall(r'\s?(\s*\S+)', f.readline().rstrip())
    
    for k in check_class:
        class_list = check_class[k]
        idx_list=[]
        for c in class_list:
            idx_list.append(classes.index(c))
        check_class_index[k]=idx_list
        
    for i in range(num_annotation):
        data = f.readline()
        
        filename = os.path.splitext(data.split(" ")[0])[0]
        filename_format = data.split(" ")[0]
        labels = re.findall(r'\s?(\s*\S+)', data.rstrip())[1:]
        
        check_dict[filename_format]=[]
        
        for k in check_class_index:
            index_list = check_class_index[k]
            check_labels = np.array(list(np.array(labels)[index_list]))
            check_dict[filename_format].extend(check_labels)
```

```python
check_class_index
```

```python
label_keys=[]
for val in check_class.values():
    label_keys.extend(val)
```

```python
check_csv = "check_celebA.csv"

with open(check_csv,'w') as f:
    label_line='filename,'
    for k in label_keys:
        label_line+="{},".format(k)
    label_line = label_line[:-1]+"\n"
    f.writelines(label_line)
    
    for filename in check_dict:
        check_list = check_dict[filename]
        
        new_line='{},'.format(filename)
        
        for i in range(len(check_list)):
            if int(check_list[i])==1:
                wstr = 1 # true 1
            else:
                wstr = 0 # false 0
            new_line+='{},'.format(wstr)
        new_line = new_line[:-1]+"\n"
        f.writelines(new_line)
```

```python
import random
```

```python
df = pd.read_csv("celebA/csv/230720/celebA_anno_mixed10000.csv")
```

```python
df
```

```python
postive_index=[]
negative_index=[]
```

```python
for i,val in enumerate(df.values):
    label_list = val[4:]
    true_check = np.any(label_list==True)
    
    if true_check:
        postive_index.append(i)
    else:
        negative_index.append(i)
```

```python
len(postive_index),len(negative_index)
```

```python
train_ratio=0.8
val_ratio=0.1
test_ratio=0.1
```

```python
train_p=[]
train_n=[]
val_p=[]
val_n=[]
test_p=[]
test_n=[]
```

```python
val_p = random.sample(postive_index,round(len(postive_index)*val_ratio))
val_n = random.sample(negative_index,round(len(negative_index)*val_ratio))
```

```python
postive_index2 = []
negative_index2 = []
```

```python
for i in postive_index:
    if i in val_p: continue
    postive_index2.append(i)
    
for i in negative_index:
    if i in val_n: continue
    negative_index2.append(i)
```

```python
len(postive_index2),len(negative_index2)
```

```python
test_p = random.sample(postive_index2,round(len(postive_index2)*test_ratio))
test_n = random.sample(negative_index2,round(len(negative_index2)*test_ratio))
```

```python
for i in postive_index2:
    if i in test_p: continue
    train_p.append(i)
    
for i in negative_index2:
    if i in test_n: continue
    train_n.append(i)
```

```python
print(len(train_p),len(train_n))
print(len(val_p),len(val_n))
print(len(test_p),len(test_n))
```
