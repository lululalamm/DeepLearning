```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
csv_path = "aihub_dataset/FaceMask/id_balanced/saved_train_id_balanced.csv"
```

```python
df = pd.read_csv(csv_path,header=None)
```

```python
df
```

```python
names = df[6]
```

```python
image_base = "aihub_dataset/FaceMask/id_balanced/images/"
```

```python
samples = random.sample(list(names.values),30)
```

```python
plt.figure(figsize=(10,10))
c=1
for s in samples:
    img_path = os.path.join(image_base,s)
    img = cv2.cvtColor(cv2.imread(img_path),cv2.COLOR_BGR2RGB)
    plt.subplot(6,5,c)
    plt.imshow(img)
    plt.axis('off')
    c+=1
```
